# Esercizio 1
a) Scrivere un codice che simuli N lanci di una moneta assumendo p = 0.5.

b) Plottare gli istogrammi per N = 100, 1000, 10000 e confrontarli con la distribuzione teorica attesa.

c) Ripetere l’esperimento nel caso di moneta truccata con p = 0 − 7.

d) Commentare i risultati ottenuti.

```python
%pip install matplotlib numpy scipy
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
def lancia(N, p, plt):
    monete = np.array([np.random.rand() for i in range(N)])
    teste = len(monete[monete < p])
    print(f"Ho lanciato {N} monete e ho ottenuto {teste} teste")

    plt.bar([0, 1], [N-teste, teste])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

N = [100, 1_000, 10_000]
for i in range(len(N)):
    lancia(N[i], 0.5, ax[i])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

N = [100, 1_000, 10_000]
for i in range(len(N)):
    lancia(N[i], 0.7, ax[i])
```

# Esercizio 2
a) Calcolare utilizzando il metodo hit or miss il seguente integrale definito: (vedi pdf)

b) Plottare il risultato dell’integrale in funzione del numero di punti generati.

c) Ripetere il conto nel caso in cui a = −1 e b = 1.

d) Quale converge più rapidamente?

```python
norm.pdf(-3)
```

```python
a, b = -3, 3
max = 1 / np.sqrt(2 * np.pi)
```

```python
N = 1_000_0

is_below_count = 0
for i in range(N):
    x = np.random.rand() * (b - a) + a
    y = np.random.rand() * max

    if y < norm.pdf(x):
        is_below_count += 1
```

```python
area = (b - a) * max
p = (is_below_count / N)
integrale = area * p
print(f"Area integrale: {integrale}")
print(is_below_count)
```
